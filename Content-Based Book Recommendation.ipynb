```python
df = pd.read_csv(r'C:\Users\acer\Downloads\cleaned_book_dataset.csv')
df.head()
```

```python
data_url = []
for i in df['ISBN']:
  url = 'https://covers.openlibrary.org/b/isbn/{}-L.jpg'.format(i)
  data_url.append(url)
df['url'] = data_url
df
```

```python
df = df[df['Description'].notna()]
df.shape
```

```python
df = df.iloc[:1000,:]
df.shape
```

```python
df['cleaned_desc'] = df['cleaned_desc'].map(str)
```

```python
df['cleaned_desc'].dtype
```

```python
df.shape
```

```python
#df.set_index('Name', inplace=True)
tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(df['cleaned_desc'].values.astype('U'))
cos_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
cos_sim
```

```python
# Set index utama di kolom 'name'
indices = pd.Series(df.index)
indices[:50]
```

```python
def recommendations(name, cos_sim = cos_sim):
    
    recommended_hotel = []
    
    # Mengambil nama hotel berdasarkan variabel indicies
    idx = indices[indices == name].index[0]

    # Membuat series berdasarkan skor kesamaan
    score_series = pd.Series(cos_sim[idx]).sort_values(ascending = False)

    # mengambil index dan dibuat 10 baris rekomendasi terbaik
    top_10_indexes = list(score_series.iloc[1:11].index)
    
    for i in top_10_indexes:
        recommended_hotel.append(list(df.index)[i])
        
    return recommended_hotel
```

```python
recommendations('The Revelation Record')
```
